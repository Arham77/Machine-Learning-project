```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
data = load_boston()
```

```python
print (data.DESCR)
```

```python
x_axis = data.data
```

```python
y_axis = data.target
```

```python
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(y_axis,bins=50)
plt.xlabel('price')
plt.ylabel('number')
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(data.data[:,5],data.target)
plt.ylabel('number of sales')
plt.xlabel('price')
plt.show()
```
